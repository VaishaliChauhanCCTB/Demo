## Classification

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
iris = load_iris()
print(iris.feature_names)
```

```python
iris = load_iris()
# Convert to DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Use describe() to summarize the data
print(iris_df.describe())
```

```python
# Load the Iris dataset
iris = load_iris()

# Convert the data into a DataFrame
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

iris_df.head(5)
```

```python
iris = load_iris()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

print(y_train.shape)
print(y_test.shape)

# Convert X_train to a DataFrame
X_train_df = pd.DataFrame(X_train, columns=iris.feature_names)

# Print the first 5 rows of the DataFrame
print(X_train_df.head())

# Create a K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris Classification')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

iris_df = pd.DataFrame(X_test, columns=iris.feature_names)
iris_df['target'] = y_test
sns.pairplot(iris_df, hue='target')
plt.show()
```

## Clustering

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

# Generate the "blobs" dataset
X, y = make_blobs(n_samples=200, centers=4, random_state=42)

X
```

```python
# Create an Agglomerative Clustering model
clustering = AgglomerativeClustering(n_clusters=4)

# Fit the model to the data
clustering.fit(X)

# Get the cluster labels
labels = clustering.labels_

# Plotting the clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering Results')
plt.show()
```
